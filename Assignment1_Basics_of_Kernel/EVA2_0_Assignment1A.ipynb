```
# import the necessary packages

import numpy as np
from urllib.request import urlopen    # This is required to open the image from the url
import cv2    
from matplotlib import pyplot as plt

# METHOD #1: OpenCV, NumPy, and urllib
# The following function takes the url as input and returns the image after processing 

def url_to_image(url):
	# download the image, convert it to a NumPy array, and then read
	# it into OpenCV format
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

```
#----------------------------------------------------------------
# define variable to hold the url of the image . 
#This variable will be the input parameter to the function 
#----------------------------------------------------------------

image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
# Import the package from colab to display the image

from google.colab.patches import cv2_imshow
print(" The Image extracted from the url")
cv2_imshow(image)
```

```
#-----------------------------------------------------------
# We will use the Canny Edge detection algorithm here to 
#find out the edges of the image 
#-----------------------------------------------------------

edges = cv2.Canny(image,100,200)
# This will display the Processed image 
cv2_imshow(edges)
```

```
#-------------------------------------------------------------------------
# The kernel here defines the type of edge the code will produce  
#The following code will produce a horizontal edge 
#-------------------------------------------------------------------------

kernel = np.float32([[-1,-1,-1],[0,0,0],[1,1,1]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#------------------------------------------------------
# The following code will produce a 45 degree edge 
#------------------------------------------------------

kernel = np.float32([[-1,1,0],[-1,0,1],[-1,0,1]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#------------------------------------------------------
# The following code will produce a 45 degree edge 
#------------------------------------------------------

kernel = np.float32([[0,-1,1],[-1,0,1],[-1,1,0]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#------------------------------------------------
# The following code will produce a blur kernel
#------------------------------------------------

kernel = np.float32([[0,-2,5],[-2,0,4],[-2,4,0]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#------------------------------------------------
# The following code will produce a sharp edge 
#------------------------------------------------

kernel = np.float32([[-1,-1,-1],[1,1,1],[3,3,3]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

```
#------------------------------------------------------
# The following code will produce an Identical edge 
#------------------------------------------------------

kernel = np.float32([[1,0,0],[0,1,0],[0,0,1]])
dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
